```python
                        #TASK 1: Data Cleaning & Preprocessing
```

```python
import pandas as pd

df = pd.read_csv("iris.csv")

# Menampilkan 5 baris pertama
df.head()
```

```python
# Melakukan pengecekan jumlah missing values per kolom
df.isnull().sum()
```

```python
# Cek jumlah duplikat
df.duplicated().sum()

# Hapus duplikat
df.drop_duplicates(inplace=True)
```

```python
# Ubah nama kolom jadi lowercase dan tanpa spasi
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
```

```python
# Simpan dataset bersih
df.to_csv("iris_cleaned.csv", index=False)
```

```python
                        # Task 2 â€“ Exploratory Data Analysis (EDA)
```

```python
import pandas as pd

df = pd.read_csv("iris_cleaned.csv")
df.head()
```

```python
# Statistik deskriptif umum
df.describe()

# Median
df.median(numeric_only=True)

# Modus
df.mode(numeric_only=True).iloc[0]

# Standar deviasi
df.std(numeric_only=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
df.hist(figsize=(10, 8), bins=15)
plt.suptitle("Distribusi Fitur Iris", fontsize=16)
plt.show()

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop(columns=["species"]))
plt.title("Boxplot Fitur Iris")
plt.show()
```

```python
# Korelasi
corr = df.drop(columns=["species"]).corr()

# Heatmap korelasi
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Korelasi Antar Fitur")
plt.savefig("heatmap_correlation.png") 
plt.show()
```

```python
                    #TASK 3: Basic Data Visualization
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("iris_cleaned.csv")

#menghitung spesies
species_count = df['species'].value_counts().reset_index()
species_count.columns = ['species', 'count']

# Buat bar chart dengan hue 
plt.figure(figsize=(6,4))
sns.barplot(data=species_count, x="species", y="count", hue="species", palette="Set2", legend=False)
plt.title("Jumlah Data per Spesies")
plt.xlabel("Spesies")
plt.ylabel("Jumlah")
plt.savefig("bar_species_count.png") 
plt.show()
```

```python
# Mengurutkan berdasarkan sepal_length untuk garis yang rapi
df_sorted = df.sort_values(by="sepal_length")

plt.figure(figsize=(8,5))
plt.plot(df_sorted["sepal_length"], df_sorted["petal_length"], marker='o', linestyle='-', color='teal')
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.grid(True)
plt.show()
```
